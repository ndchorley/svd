```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [15, 7.5]
plt.rcParams['font.size'] = 15
```

```python
n_x = 100
x = np.linspace(start=0, stop=6*np.pi, num=n_x)
s = np.sin(x)
```

```python
plt.plot(x, s)
```

```python
n_t = 10
X = np.zeros((n_x, n_t))
```

```python
t = range(n_t)
```

```python
for t_i in range(n_t):
    X[:,t_i] = np.sin(x - t_i)
```

```python
plt.contourf(X.T, cmap="inferno")
```

```python
U, S, VT = np.linalg.svd(X, full_matrices=False)
```

```python
U.shape, S.shape, VT.shape
```

```python
figure, axes = plt.subplots()
axes.semilogy(S, 'o')
```
