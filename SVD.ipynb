```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [15, 7.5]
```

```python
x = np.linspace(start=0, stop=6*np.pi, num=100)
s = np.sin(x)
```

```python
plt.plot(x, s)
```

```python
X = np.zeros((100, 10))
```

```python
t = range(10)
```

```python
for t_i in range(10):
    X[:,t_i] = np.sin(x - t_i)
```

```python
plt.contourf(X.T, cmap="inferno")
```

```python
U, S, VT = np.linalg.svd(X, full_matrices=False)
```

```python
U.shape, S.shape, VT.shape
```

```python
figure, axes = plt.subplots()
axes.semilogy(S, 'o')
```
