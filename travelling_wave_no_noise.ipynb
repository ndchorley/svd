```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [15, 7.5]
plt.rcParams['font.size'] = 15
```

```python
n_x = 100
x = np.linspace(start=0, stop=6*np.pi, num=n_x)
s = np.sin(x)
```

```python
plt.plot(x, s)
plt.xlabel('Position along slit')
plt.ylabel('Amplitude')
```

```python
n_t = 10
X = np.zeros((n_x, n_t))
```

```python
t = range(n_t)
```

```python
for t_i in range(n_t):
    X[:,t_i] = np.sin(x - t_i)
```

```python
figure, axes = plt.subplots()

contours = plt.contourf(x, t, X.T, cmap="inferno")
plt.xlabel('Position along slit')
plt.ylabel('Time')

colorbar = figure.colorbar(contours)
colorbar.ax.set_ylabel('Amplitude')
```

```python
U, S, VT = np.linalg.svd(X, full_matrices=False)
```

```python
U.shape, S.shape, VT.shape
```

```python
figure, axes = plt.subplots()
axes.semilogy(S, 'o')
axes.set_xlabel('Index of singular value')
axes.set_ylabel('Singular value')
```

```python
plt.plot(np.cumsum(S)/np.sum(S), 'o')
plt.xlabel('Index of singular value')
plt.ylabel('Energy')
```

```python
plt.subplot(3, 1, 1)
plt.plot(x, U[:, 0])

plt.subplot(3, 1, 2)
plt.plot(x, U[:, 1])
plt.ylabel('Amplitude')

plt.subplot(3, 1, 3)
plt.plot(x, U[:, 2])
plt.xlabel('Position along slit')
```
